```python
class Neural_network:
    def __init__(self,num_input,num_output,learning_rate):
        self.num_input=num_input
        self.num_output=num_output
        self.learning_rate=learning_rate
        self.wio=np.random.normal(0.0,pow(self.num_output,-0.5),(self.num_input,self.num_output))
    def train(self,inputs_list,target_list):
        inputs=np.array(inputs_list,ndmin=2).T
        targets=np.array(target_list,ndmin=2).T
        final_outputs=np.dot(self.wio.T,inputs)
        error_output=targets-final_outputs
        self.wio+=self.learning_rate*np.dot(inputs,error_output.T)
    def query(self,input_list):
        inputs=np.array(input_list,ndmin=2).T
        final_outputs=np.dot(self.wio.T,inputs)
        return final_outputs
```

```python
#创建神经网络的实例
num_input=3#输入三个值，分别为
num_outputs=1
learning_rate=0.01
network=Neural_network(num_input,num_outputs,learning_rate)
```

#### 读取数据对神经网络进行训练

```python
#读取数据准备训练
df_data=pd.read_csv("advertising.csv")
df_data=df_data.values
df1=np.array(df_data)
def normalization(data):
    data_min=data.min(axis=0)
    data_max=data.max(axis=0)
    return (data-data_min)/(data_max-data_min)
df1=normalization(df1)
cost_input=df1[:,:3]
targets=df1[:,3].reshape(-1,1)
for i in range(1000):
    network.train(cost_input,targets)
print(network.wio)
```
