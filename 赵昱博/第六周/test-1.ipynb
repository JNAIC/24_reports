```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data.txt',sep=',',names=['people','payback'])#从txt文件中读入数据，并且起名
#data.to_excel('data.xlsx')
#pd.read_excel('data.xlsx')
#print(data)
```

```python
numpy_array=data.values#创建一个numpy数组，使它的值等于data的值
#print(numpy_array)
```

```python
#把numpy数组拆了，拆成x与ｙ
x=numpy_array[:,0]
h=numpy_array[:,1]
#print(x)
#print(h)
```

```python
#定义模型
k=0
b=0
y=0
def model(k,b,x):
    y=k*x+b
    return y
```

```python
#定义损失函数，这里用均方误差
def cost(h,y):
    return 0.5*1/97*np.sum((y-h)**2)
```

```python
#定义梯度下降的函数，使用批量梯度下降，学习率是我室友随口说的一个数  
alpha=0.02
def down(y,h,x):
    global k,b
    change_k=k-alpha*1/97*np.sum((y-h)*x)
    change_b=b-alpha*1/97*np.sum((y-h))
    k=change_k
    b=change_b
    return k,b
```
